```python
import math

class complex(object):
    
    def __init__(self, real, image):
        self.real = real
        self.image = image
    def mdl(self): #module of complex number
        return math.sqrt(self.real ** 2 + self.image)
    
    def trg(self): #angle phi
        return math.acos(self.real / complex.mdl(self))
    
    def conj(self): #conjunction
        return self.real, -self.image
    
    def __add__(self, other):
        return self.real + other.real, self.image + other.image
    
    def __sub__(self, other):
        return self.real - other.real, self.image - other.image
    
    def __mul__(self, other):
        return self.real * other.real - self.image * other.image, self.image * other.real + self.real * other.image
    
    def __truediv__(self, other):
        return (self.real * other.real - self.image * other.image)/complex.mdl(other) ** 2 , (self.real * other.image + self.image * other.real)/complex.mdl(other) ** 2
    
    def __pow__(self, h): #with complex power
         return math.cos(complex.trg(self) * h.real) * math.exp(-complex.trg(self) * h.image), math.sin(complex.trg(self) * h.real) * math.exp(-complex.trg(self) * h.image)
    
    def gleichung(self): #fiding the equasion which have this root
        return('x^2 + ' + str(2 * self.real) +'x + ' + str((complex.mdl(self)) ** 2))
    
```
