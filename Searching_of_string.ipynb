```python
import timeit
#Решение в лоб

def lob(s, b): #s - исходная строка, b - шаблон
    q = str()
    for i in range(0, len(s) - len(b) + 1):
        for j in range(i, i + len(b)):
            q += s[j]
        if q == b:
            return True
        else:
            q = ""
    return False

a = str(input())
c = str(input())
print(lob(a,c))
timeit.timeit(number = 1000)
```
